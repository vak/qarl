# Usage example

```python
import numpy as np

import src as qarl
```

## Gaussian

```python
generator = qarl.generator.gaussian(mean=20, sigma=1)
environment = qarl.environment.GeneratorEnvironment(generator)
agent = qarl.agent.Agent(
    acceleration=np.ones(1) * 1.7897121877252560,
    braking=np.ones(1) * -0.0974316650691893,
)
oracle = qarl.oracle.Oracle()
game = qarl.game.Game(agent, environment, oracle)
```

```python
data = game.run_n_iterations(400)
qarl.utils.show_game_state(data)
```
