```python
import itertools
from matplotlib import pyplot as plt
from src import generator
```

```python
def show_generator(gen, title="", steps=50):
  data = list(itertools.islice(gen, steps * 100))
  plt.figure(figsize=(20,5))
  plt.subplot(1,2,1)
  plt.title(title)
  plt.xlabel("Time")
  plt.ylabel("Generator value")
  plt.plot(data[:steps:])
  plt.subplot(1,2,2)
  _ = plt.hist(data, 32)
```

```python
show_generator(generator.gaussian(), "Gaussian generator")
```

```python
show_generator(generator.gaussian_mixture(), "Gaussian mixture generator")
```

```python
show_generator(generator.sin(), "Sin generator")
```

```python
show_generator(generator.sin_gaussian(), "Noised sin generator")
```
